```python
import My_first_model as model
import numpy as np
import matplotlib.pyplot as plt

simulation = np.array(model.run_population(tmax = 300, cells=1000, kr=5, gamma_r=1/5, kp=10, gamma_p=0.4))
```

```python
fig, ax = plt.subplots(figsize = (7, 4))
plt.plot(simulation[:].T[2,:], lw = 0.1, color='red', alpha=0.2) # Multiple Cells
plt.plot(np.mean(simulation[:].T[2,:], axis=1), color='black') # Mean
plt.xlabel('Time (a.u.)')
plt.ylabel('RNA')
```
